```python
import matplotlib.pyplot as plt
import numpy as np

def polar_pos(spatial_series, neurodata_vis_spec, **kwargs):
    fig = plt.figure()
    ax = fig.add_subplot(111, polar=True)
    tt = spatial_series.timestamps[:]

    ax.plot(spatial_series.data[:]*2*np.pi/400, tt,'.', alpha=.5)
    ax.set_rorigin(-max(tt)/4)
    
    return fig
```

```python
from nwbwidgets import default_neurodata_vis_spec
import pynwb

default_neurodata_vis_spec[pynwb.base.TimeSeries] = polar_pos
```
